# Loan Prediction - Exploratory Data Analysis

This notebook performs comprehensive exploratory data analysis on the loan dataset.

## Objectives:
- Load and understand the dataset structure
- Analyze data quality (missing values, duplicates)
- Explore statistical properties
- Visualize distributions and relationships
- Analyze target variable
- Identify patterns and insights

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

print('Libraries imported successfully!')
```

## 1. Load Dataset

```python
# Load the dataset
df = pd.read_csv('../data/raw/loan_data.csv')

print(f'Dataset loaded successfully!')
print(f'Shape: {df.shape}')
print(f'Rows: {df.shape[0]}, Columns: {df.shape[1]}')
```

## 2. Dataset Overview

```python
# Display first few rows
print('First 10 rows of the dataset:')
df.head(10)
```

```python
# Dataset information
print('Dataset Information:')
df.info()
```

```python
# Column names and types
print('Column Names and Data Types:')
print(df.dtypes)
```

## 3. Data Quality Analysis

```python
# Check for missing values
print('Missing Values Count:')
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing_Count': missing_values,
    'Percentage': missing_percentage
})

missing_df[missing_df['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False)
```

```python
# Visualize missing values
plt.figure(figsize=(12, 6))
missing_data = df.isnull().sum()
missing_data = missing_data[missing_data > 0].sort_values(ascending=False)

if len(missing_data) > 0:
    missing_data.plot(kind='bar')
    plt.title('Missing Values by Column')
    plt.xlabel('Column')
    plt.ylabel('Number of Missing Values')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print('No missing values found!')
```

```python
# Check for duplicate rows
duplicates = df.duplicated().sum()
print(f'Number of duplicate rows: {duplicates}')

if duplicates > 0:
    print('\nDuplicate rows:')
    print(df[df.duplicated()])
```

## 4. Statistical Summary

```python
# Statistical summary for numerical columns
print('Statistical Summary - Numerical Columns:')
df.describe()
```

```python
# Statistical summary for categorical columns
print('Statistical Summary - Categorical Columns:')
df.describe(include=['object'])
```

## 5. Target Variable Analysis

```python
# Analyze target variable if exists
if 'Loan_Status' in df.columns:
    target_col = 'Loan_Status'
    
    print(f'Target Variable: {target_col}')
    print('\nValue Counts:')
    print(df[target_col].value_counts())
    print('\nPercentage Distribution:')
    print(df[target_col].value_counts(normalize=True) * 100)
    
    # Visualize target distribution
    fig, ax = plt.subplots(1, 2, figsize=(14, 5))
    
    # Count plot
    sns.countplot(data=df, x=target_col, ax=ax[0])
    ax[0].set_title(f'Distribution of {target_col}')
    
    # Add percentages
    total = len(df)
    for p in ax[0].patches:
        percentage = f'{100 * p.get_height() / total:.1f}%'
        x = p.get_x() + p.get_width() / 2
        y = p.get_height()
        ax[0].annotate(percentage, (x, y), ha='center', va='bottom')
    
    # Pie chart
    df[target_col].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=ax[1])
    ax[1].set_ylabel('')
    ax[1].set_title(f'{target_col} Distribution')
    
    plt.tight_layout()
    plt.show()
else:
    print('Target variable not found in dataset')
```

## 6. Numerical Features Analysis

```python
# Get numerical columns
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print(f'Numerical columns: {numerical_cols}')
print(f'Number of numerical columns: {len(numerical_cols)}')
```

```python
# Distribution plots for numerical features
if len(numerical_cols) > 0:
    n_cols = 3
    n_rows = (len(numerical_cols) + n_cols - 1) // n_cols
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
    axes = axes.flatten() if n_rows > 1 else [axes] if n_rows == 1 else axes
    
    for idx, col in enumerate(numerical_cols):
        if idx < len(axes):
            sns.histplot(df[col], kde=True, ax=axes[idx], bins=30)
            axes[idx].set_title(f'Distribution of {col}')
            axes[idx].set_xlabel(col)
            axes[idx].set_ylabel('Frequency')
    
    # Hide extra subplots
    for idx in range(len(numerical_cols), len(axes)):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.show()
```

```python
# Box plots to identify outliers
if len(numerical_cols) > 0:
    n_cols = 3
    n_rows = (len(numerical_cols) + n_cols - 1) // n_cols
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
    axes = axes.flatten() if n_rows > 1 else [axes] if n_rows == 1 else axes
    
    for idx, col in enumerate(numerical_cols):
        if idx < len(axes):
            sns.boxplot(y=df[col], ax=axes[idx])
            axes[idx].set_title(f'Box Plot of {col}')
            axes[idx].set_ylabel(col)
    
    # Hide extra subplots
    for idx in range(len(numerical_cols), len(axes)):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.show()
```

## 7. Categorical Features Analysis

```python
# Get categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
print(f'Categorical columns: {categorical_cols}')
print(f'Number of categorical columns: {len(categorical_cols)}')
```

```python
# Value counts for each categorical column
for col in categorical_cols:
    print(f'\n{col}:')
    print(df[col].value_counts())
    print(f'Unique values: {df[col].nunique()}')
```

```python
# Count plots for categorical features
if len(categorical_cols) > 0:
    n_cols = 3
    n_rows = (len(categorical_cols) + n_cols - 1) // n_cols
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
    axes = axes.flatten() if n_rows > 1 else [axes] if n_rows == 1 else axes
    
    for idx, col in enumerate(categorical_cols):
        if idx < len(axes):
            sns.countplot(data=df, x=col, ax=axes[idx])
            axes[idx].set_title(f'Count Plot of {col}')
            axes[idx].set_xlabel(col)
            axes[idx].set_ylabel('Count')
            axes[idx].tick_params(axis='x', rotation=45)
    
    # Hide extra subplots
    for idx in range(len(categorical_cols), len(axes)):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.show()
```

## 8. Correlation Analysis

```python
# Correlation matrix for numerical features
if len(numerical_cols) > 1:
    corr_matrix = df[numerical_cols].corr()
    
    plt.figure(figsize=(12, 10))
    sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
                center=0, square=True, linewidths=1)
    plt.title('Correlation Matrix')
    plt.tight_layout()
    plt.show()
    
    print('\nHighly correlated features (|correlation| > 0.7):')
    high_corr = []
    for i in range(len(corr_matrix.columns)):
        for j in range(i+1, len(corr_matrix.columns)):
            if abs(corr_matrix.iloc[i, j]) > 0.7:
                high_corr.append((corr_matrix.columns[i], 
                                corr_matrix.columns[j], 
                                corr_matrix.iloc[i, j]))
    
    if high_corr:
        for feat1, feat2, corr in high_corr:
            print(f'{feat1} - {feat2}: {corr:.3f}')
    else:
        print('No highly correlated features found')
```

## 9. Feature vs Target Analysis

```python
# Analyze relationships between features and target
if 'Loan_Status' in df.columns:
    target_col = 'Loan_Status'
    
    # Numerical features vs target
    if len(numerical_cols) > 0:
        print('Numerical Features vs Target:')
        n_cols = 3
        n_rows = (len(numerical_cols) + n_cols - 1) // n_cols
        
        fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
        axes = axes.flatten() if n_rows > 1 else [axes] if n_rows == 1 else axes
        
        for idx, col in enumerate(numerical_cols):
            if idx < len(axes):
                sns.boxplot(data=df, x=target_col, y=col, ax=axes[idx])
                axes[idx].set_title(f'{col} by {target_col}')
        
        for idx in range(len(numerical_cols), len(axes)):
            axes[idx].set_visible(False)
        
        plt.tight_layout()
        plt.show()
    
    # Categorical features vs target
    cat_cols_no_target = [col for col in categorical_cols if col != target_col]
    if len(cat_cols_no_target) > 0:
        print('\nCategorical Features vs Target:')
        n_cols = 3
        n_rows = (len(cat_cols_no_target) + n_cols - 1) // n_cols
        
        fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
        axes = axes.flatten() if n_rows > 1 else [axes] if n_rows == 1 else axes
        
        for idx, col in enumerate(cat_cols_no_target):
            if idx < len(axes):
                sns.countplot(data=df, x=col, hue=target_col, ax=axes[idx])
                axes[idx].set_title(f'{col} by {target_col}')
                axes[idx].tick_params(axis='x', rotation=45)
                axes[idx].legend(title=target_col)
        
        for idx in range(len(cat_cols_no_target), len(axes)):
            axes[idx].set_visible(False)
        
        plt.tight_layout()
        plt.show()
```
